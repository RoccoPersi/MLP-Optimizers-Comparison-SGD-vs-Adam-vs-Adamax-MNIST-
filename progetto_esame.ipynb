```python
import matplotlib.pyplot as plt
```

```python
from mlp import Network
from mlp_adam_opt import AdamOpt
from adamax import AdaMaxOpt
```

```python
import mnist_loader
```

```python
training_data, _, test_data = mnist_loader.load_data_wrapper()
```

```python
# Parametri comuni
n_epochs = 10
mini_batch_size = 32
```

```python
epoch_time_training_sgd = []

epoch_accuracy_sgd = []

net_sgd = Network([784, 60, 10],0.1)
```

```python
net_sgd.train(training_data, test_data, n_epochs, mini_batch_size)
```

```python
epoch_time_training_adam = []

epoch_accuracy_adam = []

net_adam = AdamOpt([784, 60, 10],0.001)
```

```python
net_adam.train(training_data, test_data, n_epochs, mini_batch_size)
```

```python
epoch_time_training_adamax = []

epoch_accuracy_adamax = []

net_adamax = AdaMaxOpt([784, 60, 10],0.002)
```

```python
net_adamax.train(training_data, test_data, n_epochs, mini_batch_size)
```

```python
epoch_time_training_sgd = net_sgd.get_EpochTimes()
epoch_time_training_adam = net_adam.get_EpochTimes()
epoch_time_training_adamax = net_adamax.get_EpochTimes()

epoch_accuracy_sgd = net_sgd.get_EpochAccuracy()
epoch_accuracy_adam = net_adam.get_EpochAccuracy()
epoch_accuracy_adamax = net_adamax.get_EpochAccuracy()
```

```python
import numpy as np
```

```python
print(f"Tempo Training Totale SGD : {np.sum(epoch_time_training_sgd)}")
print(f"Tempo Training Totale Adam : {np.sum(epoch_time_training_adam)}")
print(f"Tempo Training Totale AdaMax : {np.sum(epoch_time_training_adamax)}")
```

```python
print(f"Accuracy SGD : {np.mean(epoch_accuracy_sgd)*100}%")
print(f"Accuracy Adam : {np.mean(epoch_accuracy_adam)*100}%")
print(f"Accuracy AdaMax : {np.mean(epoch_accuracy_adamax)*100}%")
```

```python
# Plot dei tempi di addestramento
plt.figure(figsize=(10, 5))
plt.plot(range(1, n_epochs + 1), epoch_time_training_adam, label='Adam')
plt.plot(range(1, n_epochs + 1), epoch_time_training_sgd, label='Gradient Descent')
plt.plot(range(1, n_epochs + 1), epoch_time_training_adamax, label='AdaMax')
plt.xlabel('Epochs')
plt.ylabel('Training Time (seconds)')
plt.title('Training Time Comparison - MLP One Layer')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot della precisione
plt.figure(figsize=(10, 5))
plt.plot(range(1,n_epochs+1),epoch_accuracy_adam, label='Adam')
plt.plot(range(1,n_epochs+1),epoch_accuracy_sgd, label='Gradient Descent')
plt.plot(range(1,n_epochs+1),epoch_accuracy_adamax, label='AdaMax')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison on Test Data - MLP One Layer')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
